# Shapiro-Wilk

Podemos usar Shapiro-Wilk para ver se a distribuição da idade é normal (formato de sino e p > 0,05):

## Bibliotecas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Amostra de idades (substitua essa lista pelos dados reais da sua amostra)
idades_amostra = [25, 30, 32, 28, 35, 22, 27, 31, 29, 33]

# Realiza o teste de Shapiro-Wilk
statistic, p_value = stats.shapiro(idades_amostra)

# Gera o histograma
plt.figure(figsize=(8, 4))
plt.hist(idades_amostra, bins=10, edgecolor='black')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.title('Histograma da Amostra de Idades')
plt.grid(True)
plt.show()

# Gera o gráfico Q-Q
plt.figure(figsize=(6, 6))
stats.probplot(idades_amostra, dist='norm', plot=plt)
plt.title('Gráfico Q-Q - Amostra de Idades')
plt.grid(True)
plt.show()

# Verifica o resultado do teste
alpha = 0.05  # Nível de significância
print(f"p value = {p_value}")
print(f"Nível de significância = {alpha}")

if p_value > alpha:
    print("A amostra parece seguir uma distribuição normal (não rejeitamos a hipótese nula).")
else:
    print("A amostra não parece seguir uma distribuição normal (rejeitamos a hipótese nula).")
```
