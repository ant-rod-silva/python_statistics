# Regressão linear

Regressão, em geral, busca tratar de um valor que não se é possível consegue estimar inicialmente.

Regressão linear é definida como uma equação que visa estimar a condicional (valor esperado) de uma variável y, com base nos dados e valores de algumas outras variáveis x

Uma regressão só é chamada “linear” quando se considera que a relação da resposta às variáveis é tratada como uma função linear (Ax+b). Aqueles modelos de regressão que não podem ser traduzidos em uma função linear dos parâmetros são denominados não-lineares.

O objetivo é encontrar a melhor linha reta que descreve a relação entre essas variáveis, permitindo-nos fazer previsões com base nessa relação.

## Exemplo Prático

Neste exemplo, temos duas colunas: "Investimento_Marketing" e "Vendas_Mensais". Cada linha representa um mês, e os valores nas colunas indicam o investimento e o total de vendas para aquele mês.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Criando um dataframe de exemplo
data = {'Investimento_Marketing': [20, 25, 40, 52, 67, 72, 81],
        'Vendas_Mensais': [118, 142, 225, 260, 295, 335, 364]}

df = pd.DataFrame(data)

# Podemos ver melhor os dados usando o comando .head()
df.head()
```

Agora, vamos realizar a análise de regressão linear para prever as vendas com base no investimento.
Usaremos a biblioteca scikit-learn, que fornece uma implementação fácil de usar da regressão linear em Python:

```python
# Separando as variáveis independentes (X) e a variável dependente (y)
x = df[['Investimento_Marketing']]
y = df['Vendas_Mensais']

# Criando uma instância do modelo de regressão linear
model = LinearRegression()

# Treinando o modelo
model.fit(x.values, y.values)
```

Vamos definir um target de 63k para nosso modelo:

```python
investimento_novo = [[63]]  # target
previsao_vendas = model.predict(investimento_novo)
print("A previsão de vendas é de:", previsao_vendas[0])
```

Plotando o resultado em um gráfico

```python
# Gerando valores preditos pelo modelo ao longo do intervalo desejado
investimento_range = np.arange(0, 100).reshape(-1, 1)
vendas_preditas = model.predict(investimento_range)

# Plotando o gráfico
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Dados Originais')
plt.plot(investimento_range, vendas_preditas, color='red', label='Regressão Linear')
plt.scatter(investimento_novo, previsao_vendas, color='green', marker='x', s=100, label='Novo Investimento')
plt.xlabel('Investimento em Marketing')
plt.ylabel('Vendas Mensais')
plt.title('Regressão Linear - Investimento em Marketing vs. Vendas Mensais')
plt.legend()
plt.grid(True)
plt.show()
```
